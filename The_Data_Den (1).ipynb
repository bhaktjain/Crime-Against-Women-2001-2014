# Crimes Against Women 2001-2014(India)
About Dataset It has state-wise and district level data on the various crimes committed against women between 2001 to 2014. Crimes that included are :

- Rape
- Kidnapping and Abduction
- Dowry Deaths
- Assault on women with intent to outrage her modesty
- Insult to modesty of Women
- Cruelty by Husband or his Relatives
- Importation of Girl

# Data Cleaning/Finding Outliers

```python
safety_dataset.drop("Unnamed: 0",axis=1, inplace=True) # the column "Unnamed: 0" is of no use
safety_dataset.head() # the dataset after we drop the column
```

```python
list(safety_dataset["STATE/UT"].unique())
```

```python
list(safety_dataset["STATE/UT"].unique())
len(list(safety_dataset["STATE/UT"].unique()))
```

In India we have only 36 States/UT but as per the data it has 72 States/UT, so when we see the STATE/UT list we came to know that some of them are entered in upper Case and some in lower case. so the count has increased than actual. To clear this issue we converted all to a lower case and then also it shows 3 No. higher than the actual and while inspecting found that there is spacing issues and then fixed it to obtain the actual.

```python
safety_dataset.replace({'A&N Islands':'A & N Islands','D&N Haveli':'D & N HAVELI','Delhi UT':'DELHI'},inplace=True)
```

```python
safety_dataset["STATE/UT"]=safety_dataset["STATE/UT"].str.casefold()
```

```python
list(safety_dataset["STATE/UT"].unique())
len(list(safety_dataset["STATE/UT"].unique()))
```

```python
safety_dataset["DISTRICT"].value_counts()
```

```python
#creating a new column "total" which stores the sum of all the crimes
safety_dataset["total"]=safety_dataset["Rape"]+safety_dataset["Kidnapping and Abduction"]+safety_dataset["Dowry Deaths"]+safety_dataset["Assault on women with intent to outrage her modesty"]+safety_dataset["Insult to modesty of Women"]+safety_dataset["Cruelty by Husband or his Relatives"]+safety_dataset["Importation of Girls"]
```

```python
safety_dataset.head()
```

```python
#Year wise analysis of crime from 2001 to 2014
plt.figure(figsize=(10,5))
safety_dataset.groupby("Year")["total"].sum().plot.bar()
#The crime rate increases, as the year increases 
```

```python
# Yearwise analysis of all the crimes between 2001-2014

plt.figure(figsize=(20,12))
plt.subplot(2,4,1)
safety_dataset.groupby("Year")["Rape"].sum().plot(title="Rape")
plt.subplot(2,4,2)
safety_dataset.groupby("Year")["Kidnapping and Abduction"].sum().plot(title="Kidnapping and Abduction")
plt.subplot(2,4,3)
safety_dataset.groupby("Year")["Dowry Deaths"].sum().plot(title="Dowry Deaths")
plt.subplot(2,4,4)
safety_dataset.groupby("Year")["Cruelty by Husband or his Relatives"].sum().plot(title="Cruelty by Husband or his Relatives")
plt.subplot(2,4,5)
safety_dataset.groupby("Year")["Insult to modesty of Women"].sum().plot(title="Insult to modesty of Women")
plt.subplot(2,4,6)
safety_dataset.groupby("Year")["Assault on women with intent to outrage her modesty"].sum().plot(title="Assault on women with intent to outrage her modesty")
plt.subplot(2,4,7)
safety_dataset.groupby("Year")["Importation of Girls"].sum().plot(title="Importation of Girls")
plt.subplot(2,4,8)
safety_dataset.groupby("Year")["total"].sum().plot(title="Total No. of Crimes")
```

```python
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls']

df1=pd.DataFrame()
for i in crimes:
    df_crimes=safety_dataset.groupby(['Year'])[i].sum()
    df1[i]=df_crimes
df1
```

```python
#percentage contribution of each category of crimes
a=[]
for i in crimes:
  a.append(df1[i].sum())
a.sort()
plt.figure(figsize=(10,15))
plt.pie(a,labels=crimes,autopct='%1.2f%%',colors=['black', 'gold', 'lightskyblue', 'lightcoral','lightpink','lightcyan','lightgreen'])
plt.tight_layout()
```

```python
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls']

df_state=pd.DataFrame()
for i in crimes:
    df_state_crimes=safety_dataset.groupby(['STATE/UT'])[i].sum()
    df_state[i]=df_state_crimes
df_state["Total"]=df_state.sum(axis=1)
df_state=df_state.sort_values(by="Total",ascending=False)
df_state.reset_index()
df_state
```

```python
top_rape_by_state=df_state.sort_values(by="Rape",ascending=False)
top_3rape=top_rape_by_state.iloc[:,0:1].head(5)
plt.figure(figsize=(10,5))
plt.pie(top_3rape.Rape,labels=top_3rape.index,colors=['lightcoral','skyblue','lightgreen',"royalblue","green"],startangle=45,autopct='%1.2f%%',textprops={'fontsize': 18},shadow=True,explode=[0.01,0.01,0.01,0.01,0.01])
plt.title("Top 5 States by No. of Cases on Rape",fontdict={'fontsize' : 20})
plt.tight_layout()
```
